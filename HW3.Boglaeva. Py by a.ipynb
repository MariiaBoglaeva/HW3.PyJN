1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
df_house['delta_renovated'] = df_house['yr_renovated'] - df_house['yr_built']
df_house.loc[df_house['yr_renovated']==0,'delta_renovated']=0
df_house.head()
```

1.3 Создайте признаки года продажи, месяца продажи

```python
def convert_date(x:list):
    x = x[0:4]
    return x
```

```python
df_house['yr_sale'] = df_house['date'].apply(convert_date)
df_house.head()
```

```python
def convert_date_month(x:list):
    x = x[4:6]
    return x
```

```python

df_house['month_sale'] = df_house['date'].apply(convert_date_month)
df_house.head()
```

1.4 Удалите признаки date, zipcode, lat, long

```python
df_house = df_house.drop(columns = ['date', 'zipcode', 'lat', 'long'])
df_house.head()
```

3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
df_house.pivot_table(index = 'bedrooms', 
                    values = 'price',
                    aggfunc = 'mean').sort_values('price')
```

3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df_house.groupby('condition').agg({'price':['min','mean', 'max']})
```
